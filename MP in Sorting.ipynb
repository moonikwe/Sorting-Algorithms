```python
import random 
import matplotlib.pyplot as plt
import numpy as np

from sorting import selection_sort, bubble_sort, insertion_sort

selection_random = []
selection_presorted = []
selection_reverse = []

bubble_random = []
bubble_presorted = []
bubble_reverse = []

insertion_random = []
insertion_presorted = []
insertion_reverse = []

num_of_elements = []

for i in range (1, 100):
    generated_list = random.sample(range(0, 10000), i)
    num_of_elements.append(i)
    
    selection_random.append(selection_sort(list(generated_list)))
    selection_presorted.append(selection_sort(sorted(list(generated_list))))
    selection_reverse.append(selection_sort(sorted(list(generated_list), key=int, reverse = True)))
    
    bubble_random.append(bubble_sort(list(generated_list)))
    bubble_presorted.append(bubble_sort(sorted(list(generated_list))))
    bubble_reverse.append(bubble_sort(sorted(list(generated_list), key=int, reverse = True)))
    
    insertion_random.append(insertion_sort(list(generated_list)))
    insertion_presorted.append(insertion_sort(sorted(list(generated_list))))
    insertion_reverse.append(insertion_sort(sorted(list(generated_list), key=int, reverse = True)))    

#########################################################################################################################
plt.plot(num_of_elements, selection_random, 'm-', label = "Selection Sort")
plt.plot(num_of_elements, bubble_random, 'r-', label = "Bubble Sort")
plt.plot(num_of_elements, insertion_random, 'y-', label = "Insertion Sort")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.axis([0, 50, 0, 500])
plt.xlabel('Number of array elements')
plt.ylabel('No. of Comparisons')
plt.title('Figure 1.1 O(n^2) Algorithms (Random Numbers)', fontsize=14, fontweight='bold')
plt.show()
#########################################################################################################################
plt.plot(num_of_elements, selection_presorted, 'm-', label = "Selection Sort")
plt.plot(num_of_elements, bubble_presorted, 'r-', label = "Bubble Sort")
plt.plot(num_of_elements, insertion_presorted, 'y-', label = "Insertion Sort")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.axis([0, 60, 0, 500])
plt.xlabel('Number of array elements')
plt.ylabel('No. of Comparisons')
plt.title('Figure 1.2 O(n^2) Algorithms (Presorted)', fontsize=14, fontweight='bold')
plt.show()
#########################################################################################################################
plt.plot(num_of_elements, selection_reverse, 'm-', label = "Selection Sort")
plt.plot(num_of_elements, bubble_reverse, 'r-', label = "Bubble Sort")
plt.plot(num_of_elements, insertion_reverse, 'y-', label = "Insertion Sort")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.axis([0, 50, 0, 500])
plt.xlabel('Number of array elements')
plt.ylabel('No. of Comparisons')
plt.title('Figure 1.3 O(n^2) Algorithms (Reverse Sorted)', fontsize=14, fontweight='bold')
plt.show()
```

## Bucket Sort

```python
import random 
import matplotlib.pyplot as plt
import numpy as np

from bucket_sort import bucket_sort

bucket_random_1 = []
bucket_presorted_1 = []
bucket_reverse_1 = []

bucket_random_half = []
bucket_presorted_half = []
bucket_reverse_half = []

bucket_random_n = []
bucket_presorted_n = []
bucket_reverse_n = []

bucket_random_ten = []
bucket_presorted_ten = []
bucket_reverse_ten = []

num_of_elements = []

for i in range (1, 50):
    generated_list = random.sample(range(0, 1000), i)
    num_of_elements.append(i)
    
    bucket_random_1.append(bucket_sort(list(generated_list), 1))
    bucket_presorted_1.append(bucket_sort(sorted(list(generated_list)), 1))
    bucket_reverse_1.append(bucket_sort(sorted(list(generated_list), key=int, reverse = True), 1))
    
    bucket_random_half.append(bucket_sort(list(generated_list), i/2))
    bucket_presorted_half.append(bucket_sort(sorted(list(generated_list)), i/2))
    bucket_reverse_half.append(bucket_sort(sorted(list(generated_list), key=int, reverse = True), i/2))
    
    bucket_random_n.append(bucket_sort(list(generated_list), i))
    bucket_presorted_n.append(bucket_sort(sorted(list(generated_list)), i))
    bucket_reverse_n.append(bucket_sort(sorted(list(generated_list), key=int, reverse = True), i))
    
    bucket_random_ten.append(bucket_sort(list(generated_list), 10))
    bucket_presorted_ten.append(bucket_sort(sorted(list(generated_list)), 10))
    bucket_reverse_ten.append(bucket_sort(sorted(list(generated_list), key=int, reverse = True), 10))

plt.plot(num_of_elements, bucket_random_1, 'm-', label = "1")
plt.plot(num_of_elements, bucket_random_half, 'r-', label = "n/2")
plt.plot(num_of_elements, bucket_random_n, 'y-', label = "n")
plt.plot(num_of_elements, bucket_random_ten, 'b-', label = "10")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.axis([0, 10, 0, 1000])
plt.xlabel('Number of array elements')
plt.ylabel('No. of Comparisons')
plt.title('Figure 3.1 Bucket Sort with random numbers', fontsize=14, fontweight='bold')
plt.show()

plt.plot(num_of_elements, bucket_presorted_1, 'm-', label = "1")
plt.plot(num_of_elements, bucket_presorted_half, 'r-', label = "n/2")
plt.plot(num_of_elements, bucket_presorted_n, 'y-', label = "n")
plt.plot(num_of_elements, bucket_presorted_ten, 'b-', label = "10")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.axis([0, 10, 0, 1000])
plt.xlabel('Number of array elements')
plt.ylabel('No. of Comparisons')
plt.title('Figure 3.2 Bucket Sort with preorted numbers', fontsize=14, fontweight='bold')
plt.show()

plt.plot(num_of_elements, bucket_reverse_1, 'm-', label = "1")
plt.plot(num_of_elements, bucket_reverse_half, 'r-', label = "n/2")
plt.plot(num_of_elements, bucket_reverse_n, 'y-', label = "n")
plt.plot(num_of_elements, bucket_reverse_ten, 'b-', label = "10")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.axis([0, 10, 0, 1000])
plt.xlabel('Number of array elements')
plt.ylabel('No. of Comparisons')
plt.title('Figure 3.3 Bucket Sort with reverse sorted numbers', fontsize=14, fontweight='bold')
plt.show()
```

## Radix Sort

```python
import random 
import matplotlib.pyplot as plt
import numpy as np

from radix_insertion import radix_insertion
from radix_count import radix_count

radix_insertion_random = []
radix_insertion_presorted = []
radix_insertion_reverse = []

radix_count_random = []
radix_count_presorted = []
radix_count_reverse = []

num_of_elements = []

for i in range (1, 100):
    generated_list = random.sample(range(0, 10000), i)
    num_of_elements.append(i)
    
    radix_insertion_random.append(radix_insertion(list(generated_list)))
    radix_insertion_presorted.append(radix_insertion(sorted(list(generated_list))))
    radix_insertion_reverse.append(radix_insertion(sorted(list(generated_list), key=int, reverse = True)))
    
    radix_count_random.append(radix_count(list(generated_list)))
    radix_count_presorted.append(radix_count(sorted(list(generated_list))))
    radix_count_reverse.append(radix_count(sorted(list(generated_list), key=int, reverse = True)))

plt.plot(num_of_elements, radix_insertion_random, 'm-', label = "Radix with Counting Sort")
plt.plot(num_of_elements, radix_count_random, 'r-', label = "Radix with Insertion Sort")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.axis([0, 100, 0, 50000])
plt.xlabel('Number of array elements')
plt.ylabel('No. of array element access')
plt.title('Figure 4.1 Radix Sort (Random Numbers)', fontsize=14, fontweight='bold')
plt.show()

plt.plot(num_of_elements, radix_insertion_presorted, 'm-', label = "Radix with Counting Sort")
plt.plot(num_of_elements, radix_count_presorted, 'r-', label = "Radix with Insertion Sort")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.axis([0, 100, 0, 50000])
plt.xlabel('Number of array elements')
plt.ylabel('No. of array element access')
plt.title('Figure 4.2 Radix Sort (Presorted)', fontsize=14, fontweight='bold')
plt.show()

plt.plot(num_of_elements, radix_insertion_reverse, 'm-', label = "Radix with Counting Sort")
plt.plot(num_of_elements, radix_count_reverse, 'r-', label = "Radix with Insertion Sort")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.axis([0, 100, 0, 50000])
plt.xlabel('Number of array elements')
plt.ylabel('No. of array element access')
plt.title('Figure 4.3 Radix Sort (Reverse)', fontsize=14, fontweight='bold')
plt.show()
```

## Merge Sort vs Quick Sort

```python
import random 
import matplotlib.pyplot as plt
import numpy as np

from merge_sort import merge_sort
from quick_sort import quick_sort

merge_random = []
quick_random = []

num_of_elements = []

for i in range (1, 100):
    generated_list = random.sample(range(0, 10000), i)
    num_of_elements.append(i)
    
    merge_random.append(merge_sort(list(generated_list)))
    quick_random.append(quick_sort(list(generated_list)))
    
plt.plot(num_of_elements, merge_random, 'm-', label = "Merge Sort")
plt.plot(num_of_elements, quick_random, 'r-', label = "Quick Sort")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.axis([0, 50, 0, 500])
plt.xlabel('Number of array elements')
plt.ylabel('No. of Comparisons')
plt.title('Figure 5. Merge Sort vs Quick Sort on Random Numbers', fontsize=14, fontweight='bold')
plt.show()
```
